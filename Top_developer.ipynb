```python
import pandas as pd
df=pd.DataFrame(columns=["Title","Content","Employee","Stipend","Salary","Country"])
df
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--window-size=1200x600')
chrome_options.add_argument('--no-sandbox')  # Add this line
chrome_options.add_argument('--disable-dev-shm-usage')  # Add this line
chrome_options.binary_location = '/usr/bin/chromium-browser'  # Set the correct path
driver = webdriver.Chrome(options=chrome_options)
url_template = "https://www.topdevelopers.co/directory/big-data-analytics-companies?page={}"

for i in range(1,27):
    page_url = url_template.format(i)
    driver.get(page_url)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    press=soup.find(class_="col-xl-12 col-lg-12 col-md-12 col-sm-12 col-12 comp_listing pt-4")
    row=press.find_all(class_="comp_main_box shadow-sm mb-2 bg-white card_box rounded")
    for comp_main_box in row:
        data=comp_main_box.find(class_="row")
        title=data.find("span",itemprop="name").text

        content_url=data.find("a",class_="btn btn-view-website text-capitalize visit_site",href=True)["href"]
        driver1 = webdriver.Chrome(options=chrome_options)
        driver1.get(content_url)
        page_source = driver1.page_source
        soup1 = BeautifulSoup(page_source, 'html.parser')
        content= soup1.find(class_="col-xl-12 col-lg-12 col-md-12 col-sm-12 col-12 text-justify mb-1 cmp_content_none desc").text.strip()

        employee=data.find("div",class_="col-12 col-sm-12 col-md-12 col-lg-12 col-xl-12 order-0 order-md-1").text.strip()
        stipend=data.find("div",class_="col-12 col-sm-12 col-md-12 col-lg-12 col-xl-12 order-2 order-md-3").text.strip()
        salary=data.find("div",class_="col-12 col-sm-12 col-md-12 col-lg-12 col-xl-12 order-3 order-md-4").text.strip()
        country=data.find("div",class_="col-12 col-sm-12 col-md-12 col-lg-12 col-xl-12 order-4 order-md-5").text.strip()

        l=[title,content,employee,stipend,salary,country]
        df.loc[len(df)]=l
        df.drop_duplicates(inplace=True)

        driver1.quit()

driver.quit()
```
